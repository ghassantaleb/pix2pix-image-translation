```python
import tensorflow as tf
import tensorflow_datasets as tfds

# Load the fashion_mnist dataset
dataset, info = tfds.load('fashion_mnist', with_info=True, as_supervised=True)

# Split the dataset into training and testing
train_dataset, test_dataset = dataset['train'], dataset['test']

# Normalize and resize the images
def preprocess_image(image, label):
    image = tf.image.convert_image_dtype(image, tf.float32)  # Normalize to [0, 1]
    image = tf.image.resize(image, [128, 128])  # Resize to 128x128
    return image, label  # For simplicity, using the same image as target

# Prepare the dataset for training
train_dataset = train_dataset.map(preprocess_image)
train_dataset = train_dataset.shuffle(1000).batch(1).prefetch(tf.data.experimental.AUTOTUNE)

```

```python
def build_generator():
    inputs = tf.keras.Input(shape=[128, 128, 1])  # Input shape for the generator

    # Encoder
    x = inputs # Assign input to x initially
    down_stack = [
        tf.keras.layers.Conv2D(64, (4, 4), strides=2, padding='same', activation='relu'),  # (64, 64, 64)
        tf.keras.layers.Conv2D(128, (4, 4), strides=2, padding='same', activation='relu'),  # (32, 32, 128)
        tf.keras.layers.Conv2D(256, (4, 4), strides=2, padding='same', activation='relu'),  # (16, 16, 256)
        tf.keras.layers.Conv2D(512, (4, 4), strides=2, padding='same', activation='relu'),  # (8, 8, 512)
    ]
    # Apply each layer in down_stack to x and update x
    for down in down_stack:
        x = down(x)

    # Bottleneck
    # Apply Conv2D layer to x which now holds the output of the last down_stack layer
    bottleneck = tf.keras.layers.Conv2D(512, (4, 4), padding='same', activation='relu')(x)

    # Decoder
    up_stack = [
        tf.keras.layers.Conv2DTranspose(256, (4, 4), strides=2, padding='same', activation='relu'),  # (16, 16, 256)
        tf.keras.layers.Conv2DTranspose(128, (4, 4), strides=2, padding='same', activation='relu'),  # (32, 32, 128)
        tf.keras.layers.Conv2DTranspose(64, (4, 4), strides=2, padding='same', activation='relu'),   # (64, 64, 64)
    ]
    # Apply each layer in up_stack to x and update x
    for up in up_stack:
        x = up(x)

    # Output layer
    last = tf.keras.layers.Conv2DTranspose(1, (4, 4), strides=2, padding='same', activation='sigmoid')  # (128, 128, 1)

    x = last(x)
    return tf.keras.Model(inputs=inputs, outputs=x)

    # Connect the layers
    x = inputs
    for down in down_stack:
        x = down(x)

    for up in up_stack:
        x = up(x)

    x = last(x)
    return tf.keras.Model(inputs=inputs, outputs=x)
```

```python
def build_discriminator():
    inputs = tf.keras.Input(shape=[128, 128, 1])  # Input shape for the discriminator
    x = tf.keras.layers.Conv2D(64, (4, 4), strides=2, padding='same')(inputs)  # (64, 64, 64)
    x = tf.keras.layers.LeakyReLU()(x)

    x = tf.keras.layers.Conv2D(128, (4, 4), strides=2, padding='same')(x)  # (32, 32, 128)
    x = tf.keras.layers.LeakyReLU()(x)

    x = tf.keras.layers.Conv2D(256, (4, 4), strides=2, padding='same')(x)  # (16, 16, 256)
    x = tf.keras.layers.LeakyReLU()(x)

    x = tf.keras.layers.Conv2D(512, (4, 4), strides=2, padding='same')(x)  # (8, 8, 512)
    x = tf.keras.layers.LeakyReLU()(x)

    last = tf.keras.layers.Conv2D(1, (4, 4), padding='same')(x)  # (8, 8, 1)

    return tf.keras.Model(inputs=inputs, outputs=last)
```

```python
loss_object = tf.keras.losses.BinaryCrossentropy(from_logits=True)

def generator_loss(generated_output):
    return loss_object(tf.ones_like(generated_output), generated_output)

def discriminator_loss(real_output, generated_output):
    real_loss = loss_object(tf.ones_like(real_output), real_output)
    generated_loss = loss_object(tf.zeros_like(generated_output), generated_output)
    return real_loss + generated_loss
```

```python
# Initialize the models
generator = build_generator()
discriminator = build_discriminator()

# Define optimizers
generator_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)
discriminator_optimizer = tf.keras.optimizers.Adam(2e-4, beta_1=0.5)

# Training function (as defined previously)
@tf.function
def train_step(images):
    noise = tf.random.normal([tf.shape(images[0])[0], 128, 128, 1])  # Noise vector for generator
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        # Generate images
        generated_images = generator(images[0], training=True)

        # Discriminator output
        real_output = discriminator(images[0], training=True)
        fake_output = discriminator(generated_images, training=True)

        # Calculate losses
        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    # Calculate gradients
    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    # Apply gradients
    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))

# Training loop (as defined previously)
EPOCHS = 50

for epoch in range(EPOCHS):
    for image_batch in train_dataset:
        train_step(image_batch)

    print(f'Epoch {epoch + 1}/{EPOCHS} completed.')
```
